```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generuj_trajektorie_kompensowanego_procesu_poissona(T, lambda_):
    t = 0
    I = 0
    trajektoria = []

    while True:
        U = np.random.rand()
        t -= (1 / lambda_) * np.log(U)
        
        if t > T:
            break
        
        I += 1
        trajektoria.append(t)

    return np.array(trajektoria)
```

```python
T = 10  # Czas końcowy
lambda_ = 2  # Intensywność procesu Poissona

# Generowanie trajektorii kompensowanego procesu Poissona
trajektoria = generuj_trajektorie_kompensowanego_procesu_poissona(T, lambda_)
print("Wygenerowana trajektoria:", trajektoria)
```

```python
lambdy = [2, 4]  # Intensywności procesu Poissona

plt.figure(figsize=(10, 6))

for lambda_ in lambdy:
    trajektoria = generuj_trajektorie_kompensowanego_procesu_poissona(T, lambda_)
    plt.step(trajektoria, np.arange(len(trajektoria)), where='post', label=f'lambda={lambda_}')
    
    # Dodanie przewidywanego przebiegu
    oczekiwany_przebieg = np.linspace(0, T, 100)
    plt.plot(oczekiwany_przebieg, lambda_ * oczekiwany_przebieg, '--', label=f'Przewidywany przebieg dla lambda={lambda_}')

plt.xlabel('Czas')
plt.ylabel('N(t)')
plt.title('Trajektoria kompensowanego procesu Poissona')
plt.legend()
plt.grid(True)
plt.show()
```
